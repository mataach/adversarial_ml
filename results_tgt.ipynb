```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

pd.options.display.max_rows = 100
plt.rcParams['figure.dpi'] = 100

pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', 300)

dir_figures = './figures/'
dir_results = './results/'
```

```python
df_cifar = pd.DataFrame()
df_cifar = pd.read_csv(dir_results + 'eval_tgt-whitebox-cifar10.csv', index_col = 0)
df_svhn = pd.DataFrame()
df_svhn = pd.read_csv(dir_results + 'eval_tgt-whitebox-svhn.csv', index_col = 0)
```

```python
model_names = list(df_cifar['model_name'].unique())
```

```python
temp_fp = df_cifar[(df_cifar['model_name'] == 'resnet18') & (df_cifar['eps'] == 10)][['model_name', 'atk_param', 'UER', 'tgt_success']]
    
for model in model_names:
    temp = df_cifar[(df_cifar['model_name'] == model) & (df_cifar['eps'] == 10)][['model_name', 'atk_param', 'UER', 'tgt_success']]
    plt.figure(figsize = (6, 3))
    
    plt.title(model)
    plt.xticks(range(10))
    plt.grid(axis = 'y', alpha = 0.5)
    plt.xlabel('Label')
    plt.ylabel('Targeted Success')
    plt.ylim([0, 1])
    plt.bar(temp['atk_param'], temp['tgt_success'], alpha = 0.5)
    plt.bar(temp_fp['atk_param'], temp_fp['tgt_success'], alpha = 0.5, label = 'Full Precision')
    plt.legend()
    
    plt.savefig(dir_figures + 'tgt-%s-cifar10.png' % model, dpi = 300, bbox_inches='tight')
    
```

```python
temp_fp = df_svhn[(df_svhn['model_name'] == 'resnet18') & (df_svhn['eps'] == 10)][['model_name', 'atk_param', 'UER', 'tgt_success']]
    
for model in model_names:
    temp = df_svhn[(df_svhn['model_name'] == model) & (df_svhn['eps'] == 10)][['model_name', 'atk_param', 'UER', 'tgt_success']]
    plt.figure(figsize = (6, 3))
    
    plt.title(model)
    plt.xticks(range(10))
    plt.grid(axis = 'y', alpha = 0.5)
    plt.xlabel('Label')
    plt.ylabel('Targeted Success')
    plt.ylim([0, 1])
    plt.bar(temp['atk_param'], temp['tgt_success'], alpha = 0.5)
    plt.bar(temp_fp['atk_param'], temp_fp['tgt_success'], alpha = 0.5, label = 'Full Precision')
    plt.legend()
    
    plt.savefig(dir_figures + 'tgt-%s-svhn.png' % model, dpi = 300, bbox_inches='tight')
    
```
